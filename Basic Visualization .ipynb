```python
import matplotlib.pyplot as plt
```

```python
x = [6,7,8,9]
y = [1,2,3,4]
plt.plot(x,y)
```

```python
plt.bar(x,y)
```

```python
plt.scatter(x,y)
plt.xlabel('x-values')
plt.ylabel('y-values')
plt.show()
```

```python
## Subplots ##
## it takes 3 arguments i,e rows, columns and index ##
```

```python
plt.subplot(2,3,1)
plt.plot(x,y)
plt.subplot(2,3,2)
plt.scatter(x,y)
plt.subplot(2,3,3)
plt.bar(x,y)
plt.show()
```
